```python
import pandas as pd
import sqlite3
import plotly.express as px
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# 📌 1. Carga de datos y revisión de columnas
df = pd.read_csv("dataset.csv")
print("📊 Columnas del DataFrame:", df.columns)

# 📌 2. Preprocesamiento de datos
df.ffill(inplace=True)  # Manejo de valores nulos

# Convertir variables categóricas a numéricas
categorical_columns = df.select_dtypes(include=['object']).columns
encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])

# Verificar la existencia de la columna objetivo
target_column = "target" if "target" in df.columns else df.columns[-1]
X = df.drop(target_column, axis=1)
y = df[target_column]

# 📌 3. Manejo de desbalanceo de clases con SMOTE
class_counts = pd.Series(y).value_counts()
if class_counts.min() > 1:  # Aplicar SMOTE solo si todas las clases tienen al menos 2 muestras
    smote = SMOTE(sampling_strategy="auto")
    X_resampled, y_resampled = smote.fit_resample(X, y)
else:
    print("⚠️ Advertencia: Algunas clases tienen menos de 2 muestras, no se puede aplicar SMOTE.")
    X_resampled, y_resampled = X, y

# 📌 4. División en entrenamiento y prueba
split_size = min(0.3, max(0.1, 2 / len(y_resampled)))  # Ajuste dinámico del tamaño de prueba
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=split_size, random_state=42)

# 📌 5. Almacenamiento en SQLite y consultas SQL
conn = sqlite3.connect(":memory:")
df.to_sql("dataset", conn, index=False, if_exists="replace")

column_name = "category" if "category" in df.columns else df.columns[0]
query_avg = f"SELECT {column_name}, AVG(price) AS avg_price FROM dataset GROUP BY {column_name};"
df_avg_price = pd.read_sql(query_avg, conn)
conn.close()

# 📌 6. Entrenamiento y ajuste de hiperparámetros con KFold (evita errores de clases pequeñas)
cv_splits = min(2, len(y_train))  # Asegurar que n_splits no sea mayor que el conjunto de entrenamiento
cv_strategy = KFold(n_splits=cv_splits, shuffle=True, random_state=42)

model = xgb.XGBClassifier()
param_grid = {"max_depth": [3, 5], "learning_rate": [0.1, 0.01], "n_estimators": [100, 200]}
grid_search = GridSearchCV(model, param_grid, scoring="accuracy", cv=cv_strategy, error_score="raise")
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

# 📌 7. Evaluación del modelo
y_pred = best_model.predict(X_test)
print("\n✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📈 Reporte de Clasificación:\n", classification_report(y_test, y_pred))
print("\n🔍 Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))

# 📌 8. Visualización de importancia de características
xgb.plot_importance(best_model)
plt.show()

# 📌 9. Creación de gráficos con Plotly
fig1 = px.histogram(df, x="price", title="Distribución de Precios")
fig2 = px.scatter(df, x="price", y="sales", title="Relación Precio vs Ventas")
fig3 = px.bar(df_avg_price, x=column_name, y="avg_price", title="Precio Promedio por Categoría")

print("✅ Código ejecutado correctamente. Revisa los gráficos y la importancia de características.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# 📌 1. Cargar datos
df = pd.read_csv("dataset.csv")
print("📊 Columnas del DataFrame:", df.columns)

# 📌 2. Preprocesamiento
df.ffill(inplace=True)  # Manejo de valores nulos
categorical_columns = df.select_dtypes(include=['object']).columns
encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])

# Definir variables predictoras y objetivo
target_column = "target" if "target" in df.columns else df.columns[-1]
X = df.drop(target_column, axis=1)
y = encoder.fit_transform(df[target_column])

# 📌 3. Exploratory Data Analysis (EDA)
sns.histplot(df["price"], bins=30, kde=True)
plt.title("Distribución de Precios")
plt.show()

sns.scatterplot(x=df["price"], y=df["sales"])
plt.title("Relación Precio vs. Ventas")
plt.show()

# 📌 4. Manejo de desbalanceo con SMOTE
if min(pd.Series(y).value_counts()) > 1:
    smote = SMOTE(sampling_strategy="auto")
    X_resampled, y_resampled = smote.fit_resample(X, y)
else:
    X_resampled, y_resampled = X, y

# 📌 5. División en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# 📌 6. Entrenamiento y ajuste de hiperparámetros con XGBoost
cv_strategy = KFold(n_splits=min(2, len(y_train)), shuffle=True, random_state=42)
model = xgb.XGBClassifier()
param_grid = {"max_depth": [3, 5], "learning_rate": [0.1, 0.01], "n_estimators": [100, 200]}
grid_search = GridSearchCV(model, param_grid, scoring="accuracy", cv=cv_strategy, error_score="raise")
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_

# 📌 7. Evaluación del modelo
y_pred = best_model.predict(X_test)
print("\n✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📈 Reporte de Clasificación:\n", classification_report(y_test, y_pred))
print("\n🔍 Matriz de Confusión:\n", confusion_matrix(y_test, y_pred))

# 📌 8. Visualización de importancia de características
xgb.plot_importance(best_model)
plt.show()

# 📌 9. Guardado del modelo
import pickle
with open("xgboost_model.pkl", "wb") as model_file:
    pickle.dump(best_model, model_file)

print("✅ Modelo guardado exitosamente como 'xgboost_model.pkl'")

```
